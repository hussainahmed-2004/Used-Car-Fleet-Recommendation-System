````python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
````

````python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/139.0.0.0 Safari/537.36",

    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8",

    "Accept-Language": "en-US,en;q=0.5",

    "Accept-Encoding": "gzip, deflate, br",

    "Connection": "keep-alive",

    "Upgrade-Insecure-Requests": "1"
}
webpage=requests.get("https://www.cardekho.com/used-cars+in+new-delhi", headers=headers).text
````

````python
soup=BeautifulSoup(webpage,'lxml')
````

````python
soup.prettify()
````

````python
for i in soup.find_all('h3', class_='title'):
    print(i.text)
````

````python
for i in soup.find_all('h3', class_='title'):

    a_tag = i.find('a')

    texts = [
        text.strip()
        for text in a_tag.contents
        if isinstance(text, str) and text.strip()
    ]

    print(texts)
````

````python
for i in soup.find_all('div',class_='dotsDetails'):
    texts=[
        text.strip()
        for text in i.contents
        if isinstance(text,str) and text.strip()
    ]
    print(texts)
````

````python
model=[]
variant=[]
year=[]
Odometer_Reading=[]
fuel_type=[]
transmission=[]
price=[]
cards = soup.find_all('div', class_='cardColumn')
for card in cards:

    # ---------- TITLE ----------

    title = card.find('h3', class_='title')

    a_tag = title.find('a')

    texts = [
        text.strip()
        for text in a_tag.contents
        if isinstance(text, str) and text.strip()
    ]

    year.append(texts[0])
    model.append(texts[1])
    variant.append(texts[2])

    # ---------- DETAILS ----------

    details = card.find('div', class_='dotsDetails')

    specs = [
        text.strip()
        for text in details.contents
        if isinstance(text, str) and text.strip()
    ]

    Odometer_Reading.append(specs[0] + " " + specs[1])

    fuel_type.append(specs[2])

    transmission.append(specs[3])

    # ---------- PRICE ----------

    price_div = card.find('div', class_='priceAssured')

    price.append(price_div.text.strip())
````

````python
d={
    'Model':model,
    'Variant':variant,
    'Year':year,
    'Odometer_Reading':Odometer_Reading,
    'Fuel_type':fuel_type,
    'Transmission':transmission,
    'Price':price
}
df=pd.DataFrame(d)
````

````python
import time
final=pd.DataFrame()
for j in range(1,151):
    try:
        url="https://www.cardekho.com/used-cars+in+new-delhi/page-{}".format(j)
        webpage=requests.get(url, headers=headers).text
        soup=BeautifulSoup(webpage,'lxml')
        time.sleep(2)
        model=[]
        variant=[]
        year=[]
        Odometer_Reading=[]
        fuel_type=[]
        transmission=[]
        price=[]
        cards = soup.find_all('div', class_='cardColumn')
        for card in cards:
        
            # ---------- TITLE ----------
        
            title = card.find('h3', class_='title')
        
            a_tag = title.find('a')
        
            texts = [
                text.strip()
                for text in a_tag.contents
                if isinstance(text, str) and text.strip()
            ]
        
            year.append(texts[0])
            model.append(texts[1])
            variant.append(texts[2])
        
            # ---------- DETAILS ----------
        
            details = card.find('div', class_='dotsDetails')
        
            specs = [
                text.strip()
                for text in details.contents
                if isinstance(text, str) and text.strip()
            ]
        
            Odometer_Reading.append(specs[0] + " " + specs[1])
        
            fuel_type.append(specs[2])
        
            transmission.append(specs[3])
        
            # ---------- PRICE ----------
        
            price_div = card.find('div', class_='priceAssured')
        
            price.append(price_div.text.strip())
    
            d={
            'Model':model,
            'Variant':variant,
            'Year':year,
            'Odometer_Reading':Odometer_Reading,
            'Fuel_type':fuel_type,
            'Transmission':transmission,
            'Price':price
            }
            df=pd.DataFrame(d)
        if j % 10 == 0:
    
            print("Sleeping for 60 seconds...")
    
            time.sleep(60)

    except Exception as e:

        print(f"Page {j} failed")

        print(e)

        continue
    final = pd.concat([final, df], ignore_index=True)
````

````python
final['Price'] = final['Price'].str.replace('₹', '')
````

````python
def convert_price(price):

    if 'Lakh' in price:

        value = float(price.replace(' Lakh', ''))

        return value * 100000

    elif 'Crore' in price:

        value = float(price.replace(' Crore', ''))

        return value * 10000000

    else:

        return None
````

````python
final['Price'] = final['Price'].apply(convert_price)
````

````python
final['Car_age']=2026-final['Year']
````

````python
final['Price_per_year']=round(final['Price']/final['Car_age'],2)
````

````python
final['KM_per_year']=round(final['Odometer_Reading']/final['Car_age'],2)
````

````python
brands = [
    'Audi',
    'BMW',
    'Bentley',
    'BYD',
    'Chevrolet',
    'Citroen',
    'Datsun',
    'Fiat',
    'Ford',
    'Honda',
    'Hyundai',
    'Jaguar',
    'Jeep',
    'Kia',
    'Land Rover',
    'Lexus',
    'MG',
    'Mahindra',
    'Maruti Suzuki',
    'Mercedes-Benz',
    'Mini',
    'Nissan',
    'Porsche',
    'Renault',
    'Skoda',
    'Tata',
    'Toyota',
    'Volkswagen',
    'Volvo'
]
def extract_brand(model):

    for brand in brands:

        if model.startswith(brand):

            return brand

    return 'Other'
````

````python
final['Brand'] = final['Model'].apply(extract_brand)
````

### Handling Extreme Mileage Outliers

During exploratory data analysis, a small number of listings were found with extremely high odometer readings exceeding 1,000,000 KM.

Such values are highly unrealistic for standard passenger vehicles and may result from:
- data-entry errors,
- scraping inconsistencies,
- formatting issues,
- or corrupted marketplace listings.

These extreme outliers can significantly distort:
- scatter plots,
- statistical summaries,
- correlation analysis,
- and overall interpretation of resale price trends.

To improve data reliability and maintain analytical consistency, listings with odometer readings greater than 1,000,000 KM were removed from the dataset.

df = df[df['Odometer_Reading'] < 1000000]

````python
plt.figure(figsize=(10,6))

sns.scatterplot(
    x=df['Odometer_Reading']/1000,
    y=df['Price']/100000,
    data=df,
    alpha=0.5
)

plt.xlabel('KM Driven Thousend')

plt.ylabel('Price (Lakh ₹)')

plt.title('KM Driven vs Price')

plt.show()
````

### KM Driven vs Price Analysis

The scatter plot shows a general negative relationship between odometer reading and resale price.

As the number of kilometers driven increases, the resale value of cars tends to decrease. This is expected because higher mileage is usually associated with:
- greater vehicle wear and tear,
- increased maintenance requirements,
- reduced engine life,
- and lower buyer preference in the used-car market.

Most vehicles in the dataset are concentrated below 150,000 KM, indicating that the majority of used-car listings fall within a moderate usage range.

The visualization also reveals:
- high-priced vehicles are mostly concentrated at lower mileage levels,
- while cars with very high mileage generally belong to lower price ranges.

However, price variation still exists at similar mileage levels, suggesting that resale value is also influenced by additional factors such as:
- brand reputation,
- vehicle age,
- transmission type,
- and model segment.

The chart also helped identify a few extreme mileage outliers, which were further investigated during data cleaning.

````python
plt.figure(figsize=(8,6))

sns.boxplot(
    x='Transmission',
    y=df['Price']/100000,
    data=df
)

plt.xlabel('Transmission')

plt.ylabel('Price (Lakh ₹)')

plt.title('Transmission vs Price')

plt.show()
````

### Transmission vs Price Analysis

The boxplot compares the resale price distribution between automatic and manual transmission vehicles.

The visualization shows that automatic cars generally have higher resale prices compared to manual cars.

Key observations:
- The median resale price of automatic vehicles is noticeably higher than that of manual vehicles.
- Automatic cars also exhibit a wider price distribution, indicating the presence of both mid-range and premium luxury vehicles in this category.
- Manual vehicles are more concentrated within lower price ranges, suggesting they are more common in budget and economy segments.

A large number of high-price outliers are visible in the automatic category. These are likely associated with:
- luxury brands,
- premium SUVs,
- and high-end vehicles that predominantly use automatic transmission systems.

This pattern suggests that transmission type is strongly associated with vehicle segment and resale pricing in the used-car market.

From a buyer perspective, automatic vehicles may require a higher purchase budget but often belong to newer or premium vehicle categories.

````python
plt.figure(figsize=(10,10))

sns.boxplot(
    x=df['Price']/100000,
    y='Brand',
    data=df,orient='h'
)

plt.xlabel('Price (Lakh ₹)')

plt.ylabel('Brand Name')

plt.title('Brand vs Price Distribution')

plt.show()
````

### Brand vs Price Distribution Analysis

The boxplot compares the resale price distributions across different car brands in the used-car market.

The visualization clearly shows that premium brands such as BMW, Mercedes-Benz, Lexus, Porsche, Audi, and Land Rover have significantly higher resale price ranges compared to mass-market brands.

Key observations:
- Luxury brands exhibit higher median resale prices and wider price distributions.
- Economy and mass-market brands such as Maruti Suzuki, Hyundai, Tata, and Renault are concentrated within lower price ranges.
- Premium brands also show larger variability in pricing, likely due to broader product segmentation across luxury SUVs, sedans, and performance vehicles.
- Several extreme price outliers are visible, representing high-end flagship or imported models.

The analysis suggests that brand reputation and market positioning strongly influence resale value in the used-car market.

This indicates that resale pricing is not determined solely by vehicle age or mileage, but also by perceived brand value and segment category.
